```python
import tifffile as tf
import matplotlib.pyplot as plt

img = tf.imread('data_from_paper.ome.tif')
```

```python
plt.imshow(img[1, :, :])
```

```python
from cellpose import models, io, plot

img_single = img[0, :, :]
model = models.Cellpose(gpu=False, model_type='cyto')
masks, flows, styles, diams = model.eval([img_single], diameter=30, channels=[0,0])
```

```python
masks = masks[0]
```

```python
overlay = plot.mask_overlay(img_single, masks)
```
